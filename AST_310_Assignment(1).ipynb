```
import pandas as pd
import numpy as np
from termcolor import colored
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_excel("/content/drive/My Drive/Data for assignment.xlsx",header= 1)
data = data.drop(columns=["ID"])
#print(data.head())
name = data.columns

ses = data["SES"]
#print(ses.head())

print("1(a). Frequency Table\n")
freq_table = pd.crosstab(index= ses, columns = "count")
print(freq_table)

print("Pie Chart:- ")
ses_name = ses.unique()
#print(ses_name)
ses_value = ses.value_counts()
#print(ses_value[ses_name])

gen_name = data["Gender"].unique()
#print(gen_name)
gen_value = data["Gender"].value_counts()
#print(gen_value[gen_name])

explode = [0.0, 0.1, 0.0]

plt.pie(ses_value[ses_name],explode =explode, labels = ses_name, shadow= True, autopct = "%1.1f%%",startangle=90)
plt.title("1(b). Pie Chart for \"SES\" variable.\n")
plt.show()

print("Bar Chart :-\n")
plt.bar(ses_name, ses_value[ses_name], width=0.5, color="r")
plt.title("1(c). Bar Chart for \"SES\" variable.")
plt.show()

print("2(a). Contingency table using the variables \"SES\" and \"Gender\".\n")
conti_table = pd.crosstab(ses, data["Gender"], margins= True)
print(conti_table,"\n")

#stacked bar Diagram
print("2(b). Stacked bar diagram from the table.")
counts = data.apply(lambda x: x.value_counts() ).transpose()
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
counts.plot(ax=ax,kind='bar', stacked=True, rot=0)
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.0f}'.format(x) for x in vals])
ax.yaxis.grid(True)

ax.set_axisbelow(True)
plt.show()

```
